```python
import pandas as pd
from state import State
from string_utils import print_plan, str_to_moves

def print_best_solution(event, gem_level):
    query = df[df['event_name'] == event]
    state = State()
    state.set_event_by_name(event)
    if gem_level == 'best':
        query = query[query['time_remaining'] > 0]
        record = query.sort_values(['gem_level', 'time_remaining'], ascending = (True, False)).iloc[0]
        state.gem_level = record['gem_level']
    else:
        query = query[query['gem_level'] == gem_level]
        record = query.sort_values('time_remaining', ascending = False).iloc[0]
        state.gem_level = gem_level
    moves = str_to_moves(record['moves'])
    print(f"{state.gem_level} gem production boosts purchased")
    print(f"{event:^114}")
    print(print_plan(state, moves, include_urgency = True))
    print()

mode = 'best'
#this file is local only and not committed to the repo, you can fill in your own files with strategy.save()
df = pd.read_csv("solutions.csv", index_col = False)
for event_name in [
    'other_tower', 
    'sid_in_space', 
    'happy_time', 
    'aqua_adventure', 
    'lost_in_time', 
    'through_the_portal'
]:
    print_best_solution(event_name, mode)
```
